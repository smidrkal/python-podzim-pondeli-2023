## Creating Our First Plot

We are going to create our first plot. To do so, we need to import some libraries.

```python
# Note here we are importing Pandas as pd. We will be using pd henceforth
import pandas as pd

#Note here we are importing matplotlib.pyplot as plt. We will be using plt henceforth. 
import matplotlib.pyplot as plt
```

Before we begin, we need to tell matplotlib to behave in our environment. By default, Matplotlib will want to open up the plot in a new window. Because we are using Jupyter, we want to tell Matplotlib to render the graphs in Jupyter.

```python
# This is only because we are using Jupyter Notebook; this ensures that any plotting and figures are displayed in the Jupyter Notebook
# Otherwise it gets very iffy. 
# You ONLY use this snippet if you using Matplotlib in a Jupyter Notebook

%matplotlib inline
```

To create our first plot, we simply need to invoke the PyPlot's `plt` function `plot()`. By default, Matplotlib will create a line graph unless told otherwise. A line graph is useful when showing trends over time.

We'll explore some other plot types throughout this notebook.

```python
plt.plot()
```

Currently, our graph has no data. Let's plot some data by providing some values in our `plt.plot()`

```python
# We'll start by making a very simple plot using an array before we get into the nitty-gritty. 
# Remembering it goes X -> Y. Along the corridor and up the stairs!
plt.plot(['January', 'February', 'March', 'April' ],[8,14,23,40])
```

Okay! Now we're getting somewhere. At the moment, we don't know what this graph is telling us. Thankfully, pyplot has some functions that allow us to make this graph more descriptive. Take a look at the table below:

<table><thead><tr><th>Function</th><th>Description</th></tr></thead><tbody><tr><td><span style="font-style: normal; font-weight: 400;">plt.ylabel</span>&nbsp;</td><td>&nbsp;This function allows us to specify a label for the Y axis.</td></tr><tr><td>plt.xlabel</td><td>This function allows us to specify a label for the X axis.&nbsp;</td></tr><tr><td>plt.title&nbsp;</td><td>This functional allows us to specify a title for the plot.&nbsp;</td></tr></tbody></table>

Let's implement these to understand what the graph is showing.

```python
plt.xlabel('Months of the Year')
plt.ylabel('Number of Toys Produced')

plt.title('A Line Graph Showing the Number of Toys Produced Between September and December')

plt.plot(['September', 'October', 'November', 'December' ],[8,14,80,160])
```

That's more like it! Now we can see a visible trend in the increase in toy production. It looks like AntarctiCrafts had a busy November and December. I guess I'm not the only one who leaves my Christmas shopping last minute!

## I'll Catch You at the Bar...Graph

A bar chart is great for comparing data points. For example, comparing what the Frostling's favourite drinks are.

In this scenario, AntarctiCrafts management is trying to decide what the most popular drinks are to stock in the communal vending machine. To find out, they have asked the employees to vote on their favourite soft drinks. It is our job to plot this data for management.

Here is a snippet of the spreadsheet that we are going to be analysing:

<table><thead><tr><th>Drinks</th><th>Vote</th></tr></thead><tbody><tr><td>Holly Jolly Fizz</td><td>8</td></tr><tr><td>Mistletoe Mojito</td><td>3</td></tr><tr><td>Frostbite Frappuccino</td><td>16</td></tr></tbody></table>

Well, let's create a bar graph to find out! This time, we will read data from a spreadsheet (rather than manually providing the values).

Here's what we are going to do:
1. We will use Panda's `pd.read_csv` function to import the spreadsheet **drinks.csv** into a variable named `spreadsheet`.
2. Then, we are going to extract the columns "Drink" and "Vote" and store them in their own variables (`drinks` and `votes`).
3. Finally, we will create the bar graph to visualise the data.

```python
# Step #1.
spreadsheet = pd.read_csv('drinks.csv')

# Step #2.
drinks = spreadsheet['Drink']
votes = spreadsheet['Vote']


# Step #3.
# Here, we are going to tell pyplot the size of the figure. Find a value that improves readability
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed


# plt.bar tells pyplot to use a bar graph. the `h` means horizontal, `v` can be used for a vertical bar graph.
plt.barh(drinks, votes, color='skyblue')

# Set the labels for the graph
plt.xlabel('Number of Votes')
plt.ylabel('Name of Drink')
plt.title('A Bar Graph Showing the Employees Favourite Drinks')
plt.gca().invert_yaxis()  # Optional - invert the y-axis for better readability (most popular at the bottom)
```
